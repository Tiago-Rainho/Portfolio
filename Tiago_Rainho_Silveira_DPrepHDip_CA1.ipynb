```python
fig, ax = plt.subplots(figsize=(20,10))

missing = df.isna().sum().div(df.shape[0]).mul(100).to_frame().sort_values(by=0, ascending=False)
ax.bar(missing.index, missing.values.T[0])
plt.ylabel("Percentage missing")
plt.xticks([])
plt.show()
```

I have decided that all Columns with more than 70% missing values will be removed.

```python
columns_missing = missing[missing[0]>70]
print("There are {} columns with missing values above 70%.".format(len(columns_missing)))
```

```python
columns_to_drop = list(columns_missing.index) 
columns_to_drop.append('class')
columns_to_drop
```

```python
X = df.drop(columns_to_drop, axis=1)
y = df.loc[:,"class"]
y = pd.get_dummies(y).drop("neg",axis=1)
X.describe()
```

## Principal Component Analysis PCA  

PCA is fundamentally a dimensionality reduction algorithm that provides a method for filtering noise, is applicable to feature extraction, and is also useful as a visualization tool. (Google Colaboratory,2021)

Applying PCA we are able to identifying patterns, so we can highlight the similarities and differences between them. Its identification in data with large dimensions is difficult because its graphic representation is not viable, making visual data analysis extremely difficult.Because the focus is on the analysis of the main dimensions that characterize the data set, once the patterns in the set are identified, the number of dimensions to be analyzed can be reduced without a significant loss of information.

To begin PCA, we need to import the libraries below.

**Scikit-learn (Sklearn)** is Python's most valuable and robust machine learning library. It offers a set of essential methods for machine learning and statistical modelling, such as classification, regression, clustering, and dimensionality reduction, via a Python interface.

```python
%matplotlib inline 
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
sns.set()
```

The follow function we can verify the number of components, we need for the requested accuracy of 99.5%

```python
pca=PCA(0.995).fit(X)
pca.n_components_
```

As show above we need 3 components to reach 99.5% accuracy

```python
pca = PCA().fit(X)
plt.plot(np.cumsum(pca.explained_variance_ratio_)) 
plt.xlabel('numbers of components') 
plt.ylabel('cumulative explained variance');
```

The PCA components are shown in array form bellow so that a Machine Learning algorithm can read and process them.

```python
pca = PCA(n_components=3)
new_pca = pca.fit_transform(X)
new_pca
```

To apply dimensionality reduction let`s use PCA with three components to project the standardized aps_failure_set dataset onto the new principal component axes.

```python
pca = PCA(n_components=3) 
pca.fit(X) 
X_pca = pca.transform(X) 
print("original shape: ", X.shape) 
print("transformed shape:", X_pca.shape) 
```

```python
print(pca.components_)
```

The percentage of information for each of the three components that contain the majority of the information is listed below.

```python
explained_variance = pca.explained_variance_ratio_
explained_variance
```

In this case, we can see that using the three components, we have a 99.5% variance

```python
var = pca.explained_variance_ratio_.sum() * 100
print('Total Explained Variance: {}%'.format(round(var,3)))
```

```python
df_AFS = pd.DataFrame(data = new_pca, columns = ['P1','P2','P3'])
New_AFS = pd.concat([df_AFS, df['class']], axis = 1)
New_AFS
```

The component columns P1, P2, and P3 are the three we chose for the principal component analysis.

The df_AFS variable refers to the first dataframe that was processed by the PCA algorithm.

The variable New_AFS refers to the df_AFS after the PCA algorithm has been concatenated with the classes that have been run through it.

# CURSE OF DIMENSIONALITY

The Curse of Dimensionality, in my personal view, occurs when we have a significant number of features, which may appear to be the best choice for the model, but if they are unnecessary, it will most likely impede instead of help, because when we get to the training session, the model is likely to increase massively, and because the increase in dimensions is expected, a significant amount of data will become limited.
